```python
def preprocess(text):
    return(Counter(text.split()))

text0 = 'I am so sad I am so tired'
text1 = 'I am so tired so tired so tired'
text2 = 'I am so sad so sad so sad'
text3 = 'I am so sad and so tired'

text4 = 'I am a veloceraptor'
text5 = 'I am no loonger a veloceraptor'
text6 = 'I wish I could be a veloceraptor'
text7 = 'I am a huge veloceraptor'

distance(preprocess(text1), preprocess(text2))
```

```python
radius(preprocess(text0), preprocess(text5), [preprocess(text) for text in [text0, text1, text2, text3]])
```

```python
radius_distance(preprocess(text5), [preprocess(text) for text in [text0, text1, text2, text3]])
```

```python
from modules.single_classifier import SingleClassifier
from modules.ensamble_classifier import EnsambleClassifier
```

```python

texts = ['I am so sad I am so tired', 'I am so useless and crying and tired so tired so tired', 
         'I am so sad so sad so sad', 'I am so sad and so tired',
         
         'Wow a velocaraptor', 'She is no longer a veloceraptor',
         'I wish we could be a velocaraptor', 'velocaraptor veloceraptor veloceraptor']

ys = ['sad', 'sad', 'sad', 'sad', 'dino', 'dino', 'dino', 'dino']

profs = [preprocess(text) for text in texts]

classes = ['sad', 'dino']
cl = SingleClassifier(profs, ys, 10, classes=classes)
print(cl.forward_one(preprocess(texts[0])))

print(cl.forward_multiple([preprocess(text) for text in texts[:3]]))
```

```python
cl.forward_all()
```

```python
from modules.ulits import softmax
y_score = cl.only_distances(cl.forward_all(), arr=True)[0]  # get scores from dict
y_score = [softmax(scores) for scores in y_score]  # normilize scores
y_score
```

```python
ec = EnsambleClassifier([profs], ys, range(1, 8), classes=classes)
classifiers = ec.classifiers
classifiers[0].forward_all()
```

```python
more_texts = ['I am your veloceraptor', 'I am so sad and tired', 'this is a completely irrelevant veloceraptor', 'sad and tired sad and tired I am']
more_profs = [preprocess(text) for text in more_texts]
more_ys = ['dino', 'sad', 'dino', 'sad']

ec.forward_multiple(xs=more_profs)

ec.forward_multiple(xs=more_profs, method='weight', confidence=True)
```
